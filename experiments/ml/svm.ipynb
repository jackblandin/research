# Solve linearly separable data set using Linear SVM Kernel

```python
# Generate random x1, x2 values over [0, 1]^2 range
X = np.random.rand(100, 2)

# Make a nice looking boundary gap
X = X[(X[:, 0] + X[:, 1] > 1.1) | (X[:, 0] + X[:, 1] < .9)]

# Label each dataset with target values as x1 + x2 > 1
y = (X[:, 0] + X[:, 1] > 1).astype(int)

fig, ax = plt.subplots(1, 1, figsize=(8, 6))
C1 = X[np.where(y == 1)]
C2 = X[np.where(y == 0)]
ax.scatter(C1[:, 0], C1[:, 1], label='C1', marker='x')
ax.scatter(C2[:, 0], C2[:, 1], label='C2', marker='o')
ax.legend()
ax.set_title('Linearly separable decision boundary')
pass
```

```python
svm = SVM()
svm.fit(X, y)
svm.plot_decision_boundary(X, y)
svm.plot_discriminants(X, y)
```

```python
Xtest = np.random.rand(100, 2)
Xtest = Xtest[(Xtest[:, 0] + Xtest[:, 1] > 1.1) | (Xtest[:, 0] + Xtest[:, 1] < .9)]
ytest = (Xtest[:, 0] + Xtest[:, 1] > 1).astype(int)
preds = svm.predict(Xtest)
svm.plot_decision_boundary(Xtest, ytest)
```

# Solve radially separable data set using Polynomial SVM Kernel

```python
# Generate random x1, x2 values over [0, 1]^2 range
X = np.random.rand(20, 2)

# Make a nice looking boundary gap
X = X[
    ((X[:, 0]-.5)**2 + (X[:, 1]-.5)**2 < .09) |
    ((X[:, 0]-.5)**2 + (X[:, 1]-.5)**2 > .11)]

# Label each dataset with target values as x1 + x2 > 1
y = ((X[:, 0]-.5)**2 + (X[:, 1]-.5)**2 < .1).astype(int)

fig, ax = plt.subplots(1, 1, figsize=(8, 6))
C1 = X[np.where(y == 1)]
C2 = X[np.where(y == 0)]
ax.scatter(C1[:, 0], C1[:, 1], label='C1', marker='x')
ax.scatter(C2[:, 0], C2[:, 1], label='C2', marker='o')
ax.legend()
ax.set_title('Radially separable decision boundary')
pass
```

```python
svm = SVM(kernel='polynomial', degree=2)
svm.fit(X, y)
svm.plot_decision_boundary(X, y)
svm.plot_discriminants(X, y)
```

# Solve homework problem data using Gaussian Kernel

```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from svm import SVM


# Generate random x1, x2 values over [0, 1]^2 range
X = np.random.rand(40, 2)

y = ((X[:, 1] < .2*np.sin(10*X[:, 0]) + .3) |
     ((X[:, 1] - .8)**2 + (X[:, 0]-.5)**2 < .15**2)).astype(int)

fig, ax = plt.subplots(1, 1, figsize=(14, 10))
C1 = X[np.where(y == 1)]
C2 = X[np.where(y == 0)]
ax.scatter(C1[:, 0], C1[:, 1], label='C1', marker='x')
ax.scatter(C2[:, 0], C2[:, 1], label='C2', marker='o')
ax.legend()
pass
```

```python
svm = SVM(kernel='gaussian', radius=0.25)
svm.fit(X, y)
svm.plot_decision_boundary(X, y)
svm.plot_discriminants(X, y)
```
